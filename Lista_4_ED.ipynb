### 2. Implemente um código para verificar se uma dada árvore é, de fato, uma árvore binária.

```python
'''
Implementação 2
'''

# Código
class Node:
    def __init__(self, inputData):
        self.esquerda = None
        self.direita = None
        self.valor = inputData

# vou fazer a arvore m procedures, ao inves d abstrair
```

```python
def isArvoreBin(node, menor=None, maior=None):
    # árvore -caso base, para recursi
    if node is None:
        return True

    # Verifica se o valor do nó é menor ou igual ao valor do nó da esquerda
    if menor is not None and node.valor <= menor.valor:
        return False
    # o msmo pra direita
    if maior is not None and node.valor >= maior.valor:
        return False

    #recursiva para atingir os filhos
    return isArvoreBin(node.esquerda, menor, node) and isArvoreBin(node.direita, node, maior)
```

```python
# instancias
raiz = Node(10)
raiz.esquerda = Node(5)
raiz.direita = Node(20)
raiz.esquerda.direita = Node(15) # esse vai pra direita do nó squedo

# Verificando se a árvore é uma árvore binária de busca
print(isArvoreBin(raiz))  # Saída: True, se for uma árvore binária de busca
```

### 3. Implemente um código para fazer inclusão/exclusão em árvores binárias.

```python
'''
Implementação 3
'''

# Código

def inserirNode(raiz, chave):
    # caso base
    if raiz is None:
        return Node(chave)

    # pega a esquerda
    if chave < raiz.valor:
        raiz.esquerda = inserirNode(raiz.esquerda, chave)
    # insere na subárvore direita
    else:
        raiz.direita = inserirNode(raiz.direita, chave)

    '''fim do procdures '''
    # Retorna a raiz (inalterada)
    return raiz
```

```python
# Função para encontrar o nó com o valor mínimo (usado na exclusão)
def encontrarMinimo(no):
    atual = no

    # Encontra o nó mais à esquerda (menor valor)
    while atual.esquerda is not None:
        atual = atual.esquerda

    '''fim do procdures '''
    return atual


def excluirNo(raiz, chave):
    # Se a árvore está vazia, retorna a raiz (inalterada)
    if raiz is None:
        return raiz

    # Se a chave a ser excluída é menor que a raiz, exclui da subárvore esquerda
    if chave < raiz.valor:
        raiz.esquerda = excluirNo(raiz.esquerda, chave)
    # Se a chave a ser excluída é maior que a raiz, exclui da subárvore direita
    elif chave > raiz.valor:
        raiz.direita = excluirNo(raiz.direita, chave)
    # Se a chave é igual à raiz, este é o nó a ser excluído
    else:
        # Nó com apenas um filho ou sem filhos
        if raiz.esquerda is None:
            return raiz.direita
        elif raiz.direita is None:
            return raiz.esquerda

        # Nó com dois filhos: pega o sucessor em ordem (menor na subárvore direita)
        temp = encontrarMinimo(raiz.direita)

        # Copia o valor do sucessor em ordem para este nó
        raiz.valor = temp.valor

        # Exclui o sucessor em ordem
        raiz.direita = excluirNo(raiz.direita, temp.valor)

    # Retorna a raiz (possivelmente alterada)
    return raiz
```

```python
# Função para imprimir a árvore em ordem (para fins de teste)
def imprimirEmOrdem(no):
    if no is not None:
        imprimirEmOrdem(no.esquerda)
        print(no.valor, end=' ')
        imprimirEmOrdem(no.direita)
```

```python
# Exemplo de uso:
# Criando uma árvore binária
raiz = Node(10)
raiz = inserirNode(raiz, 5)
raiz = inserirNode(raiz, 20)
raiz = inserirNode(raiz, 15)

print("Árvore em ordem antes da exclusão:")
imprimirEmOrdem(raiz)
print()
```

```python
# Excluindo um nó da árvore binária
raiz = excluirNo(raiz, 20)

print("Árvore em ordem após a exclusão:")
imprimirEmOrdem(raiz)
print()
```
