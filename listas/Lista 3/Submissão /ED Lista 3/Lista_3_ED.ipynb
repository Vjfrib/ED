# Sorteamento/Seleção por ***contagem***

image.png


Tempo é, $O(n)$ pois itera linearmente a lista (sem ser matricial, apenas uma vez, aumentando os contadores)

```python
'''
Implementação 2
'''

# Código

# Necessário definir o range do código
def sortContagem(arr, rangeLimite):

    count = [0] * (rangeLimite + 1)  # Inicializa o array de contagem com base no range_max


    for numeros in arr:     # Preenche o array de contagem com a contagem de arr
        count[numeros] += 1

    i = 0  # Inicializa o index do array original

    # Preenche o array original com os elementos ordenados
    for valor in range(range_max + 1):
        if count[valor] > 0:
            for _ in range(count[valor]): # Loop para precheencher linearmente os qu tem contadoes repetidos
                arr[i] = valor
                i += 1
```

```python

# Exemplo de uso:
arr = [4, 2, 2, 8, 3, 3, 1]
range_max = 8  # Determina o range máximo a partir do array

sortContagem(arr, range_max)
print("Array ordenado:", arr)
```

```python
'''
Implementação 3
'''

# Código

import matplotlib.pyplot as plt
import numpy as np


# Tamanhos da entrada (n)
bigLetra = np.arange(1, 50000)  # Considerando entradas de tamanho 1 até 100

# Complexidade dos algoritmos em função de n
bubble = bigLetra**2  # O(n^2) - Bubble Sort

select = bigLetra**2  # O(n^2) - Selection Sort

insertion = bigLetra**2  # O(n^2) - Insertion Sort

merge = bigLetra * np.log2(bigLetra)  # O(n log n) - Merge Sort

quick = bigLetra * np.log2(bigLetra)  # O(n log n) - Quick Sort
```

```python

# Plotando os gráficos
plt.figure(figsize=(10, 6))
# plt.plot(bigLetra, bubble, label='bolha (O(n^2))')
# plt.plot(bigLetra, select, label='Seleção (O(n^2))')
# plt.plot(bigLetra, insertion, label='Insertion (O(n^2))')
# plt.plot(bigLetra, merge, label='Merge Sort (O(n log n))')
# plt.plot(bigLetra, quick, label='Quick Sort (O(n log n))')

plt.xlabel('Tamanho da entrada (n)')
plt.ylabel('Complexidade de tempo')
plt.title('Complexidade de Tempo dos Algoritmos de Ordenação')
plt.legend()

plt.grid(True)
plt.tight_layout()
plt.show()
```
